```python
class TestMod(nn.Module):
    d_model: int

    def setup(self):
        self.l1 = nn.Dense(self.d_model)  

    def __call__(self, x):
        x = self.l1(x.T).T
        return x
```

```python
net = TestMod(d_model=20)
```

```python
inp = np.arange(30).reshape((10, 3))
inp
```

```python
rng, rng_ = jax.random.split(jax.random.PRNGKey(0))
params = net.init(rng_, inp)
params
```

```python
net.apply(params, inp)
```
